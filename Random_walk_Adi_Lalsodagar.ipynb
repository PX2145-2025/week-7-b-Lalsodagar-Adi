# Introduction
Random processes are very common in physics. When dealing with large numbers of molecules, such as in gases, we donâ€™t describe the trajectory of every single particle ($10^{23}$ particles is too much to handle) but often use probabilities to determine the average behaviour. Here, we look at some illustrative examples of using random numbers in physical applications.

## Random walks
The random walk problem is a classic example which, despite its simplicity, describes very well the basic physical phenomenon of _diffusion_. Diffusion is a process where par- ticles which are initially concentrated in a region in space spread out with time. For example, a drop of ink in water will gradually spread to the whole volume of water, even if the water is perfectly still. This is because the individual molecules (both the ink and the water) (i) have random directions and speeds, which depend on the temperature, and (ii) also collide with each other. It turns out that these two ingredients can be modelled very easily.

## The method
We first look at a random walk in 1-D for a single particle. We start the particle off at the origin of the x-axis and assume that it moves unit distance in each time step (the molecular speed is thus 1 unit). At the end of each time step we set the direction to left or right with equal probability, mimicking the effect of collisions. You can easily write the python code, but for flexibility of later python coding, I show here how to define a function, called `walk(steps)`, which is passed a single number steps and returns a vector of that length, which contains all the positions of the random walk.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x
````

IMPORTANT: Make sure you understand how this function is used, and how it works. The rest of this worksheet is based on using it. You can also use the vectorised version below.

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    Vectorised method.
    '''
    x = np.random.choice([-1, 1], size=steps) # random -1 or 1
    x = np.cumsum(x) # cumulative sum
    return x
````

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed
````

````python
def walk(steps):
    '''
    steps is the length of the 1-D random walk vector returned
    '''
    x = np.zeros(steps, int)  # initialise the array of integers, using the 'steps' parameter
    x[0] = 0  # start at the origin
    for i in range(1, len(x)):
        rnd = random()  # rnd selected from [0,1)
        # step left or right randomly ...
        if (rnd<0.5):
            x[i] = x[i-1] + 1  # steps right one unit
        else:
            x[i] = x[i-1] - 1 # steps left one unit
    return x

# Part 1: Defining the variable for 1000 steps
num_steps = 1000

# Part 1: Plotting a single random walk for 1000 steps
positions_single = walk(num_steps)

plt.figure()
plt.plot(np.arange(num_steps), positions_single)
plt.xlabel("Step Number (i)", fontsize = 12)
plt.ylabel("Position (x)", fontsize = 12)
plt.title("Single 1D Random Walk", fontsize = 12)
plt.grid(True)
plt.show()
````

````python
# Part 2: Now creating a for loop to plot 100 random walks for 1000 steps on the same graph
num_walks = 100

plt.figure()
for _ in range(num_walks):
    positions_multiple = walk(num_steps)
    plt.plot(np.arange(num_steps), positions_multiple, linewidth = 1, alpha=0.6) # alpha for better visibility of overlapping lines

plt.xlabel("Step Number (i)", fontsize = 12)
plt.xlim(0,1000)
plt.ylabel("Position (x)", fontsize = 12)
plt.title(f"{num_walks} Random Walks", fontsize = 15)
plt.grid(alpha = 0.5)
plt.show()
````

 2. [2] Calculate (plot if you wish) the steps for 500 walkers. We define _for each step_ a "mean position": the average of the positions of all walkers at each step. We also define _for each step_ a "mean squared position": the average of the squares of the positions of all walkers (**not** the square of the average) at each step. Those two quantities, the "mean position" and the "mean squared position" are standard quantities in random walks.
````python
num_steps = 1000  # Number of steps for each walk
num_walkers = 10  # Number of walkers

# Initialising a 2D array to store the trajectories of N particles
particles = np.zeros((num_steps, num_walkers), int)

# Populating the particles array with random walks
for i in range(num_walkers):
    particles[:, i] = walk(num_steps)

# Plotting all 10 walkers on the same plot

for i in range(num_walkers):
    plt.plot(np.arange(num_steps), particles[:, i], linewidth=1, alpha=0.7)

plt.xlabel("Step Number (i)", fontsize=12)
plt.xlim(0,1000)
plt.ylabel("Position (x)", fontsize=12)
plt.title(f"{num_walkers} Random Walks (1000 steps each)", fontsize=15)
plt.grid(alpha=0.5)
plt.show()
````

````python
# Part 2: Now plotting for 500 random walks:
# Only difference this time is num_walkers = 500

num_steps = 1000  # Number of steps for each walk
num_walkers = 500  # Number of walkers

# Initialising a 2D array to store the trajectories of N particles
particles = np.zeros((num_steps, num_walkers), int)

# Populating the particles array with random walks
for i in range(num_walkers):
    particles[:, i] = walk(num_steps)

# Plotting all 500 walkers on the same plot

for i in range(num_walkers):
    plt.plot(np.arange(num_steps), particles[:, i], linewidth=1, alpha=0.7, label=f'Walker {i+1}')

plt.xlabel("Step Number", fontsize=12)
plt.xlim(0,1000)
plt.ylabel("Position (x)", fontsize=12)
plt.title(f"{num_walkers} Random Walks (1000 steps each)", fontsize=15)
plt.grid(alpha=0.5)
plt.show()
````

````python
# Calculate and print the "mean position" and "mean squared position" at the last step.

# Mean position at each step: average of positions of all walkers at each step.
mean_positions_per_step = np.mean(particles, axis=1)

# Mean squared position at each step: average of the squares of the positions of all walkers at each step.
mean_squared_positions_per_step = np.mean(particles**2, axis=1)

# For Part 2 of Exercise 2: Print out these quantities at the last step.
last_step_mean_position = mean_positions_per_step[-1]
last_step_mean_squared_position = mean_squared_positions_per_step[-1]

print(f"Mean position at the last step: {last_step_mean_position:.4f}")
print(f"Mean squared position at the last step: {last_step_mean_squared_position:.4f}")
````

````python
# Part 3: Calculate the mean position at each step, and plot it.
# The 'mean_positions_per_step' was already calculated in the previous cell.
# It contains the average of the positions of all walkers at each step.

print(mean_positions_per_step)

#Now plotting:

plt.figure()
plt.plot(np.arange(num_steps), mean_positions_per_step, linewidth = 1)
plt.xlabel("Step Number", fontsize=12)
plt.ylabel("Mean Position", fontsize=12)
plt.title("Mean Position of Walkers at Each Step", fontsize=15)
plt.grid(alpha=0.5)
plt.show()
````

````python
# Part 4: Calculating the mean square position at each step, and plotting it.
print(mean_squared_positions_per_step) # I had already defined this in a previous question

# Now I plot this, as a function of the number of steps
plt.figure()
plt.plot(np.arange(num_steps), mean_squared_positions_per_step, linewidth = 1)
plt.xlabel("Step Number", fontsize=12)
plt.ylabel("Mean square Position", fontsize=12)
plt.title("Mean Square Position of Walkers at Each Step", fontsize=15)
plt.grid(alpha=0.5)
plt.show()
````
